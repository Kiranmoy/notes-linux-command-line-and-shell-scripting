# Sorting your views line-by-line

**`sort`**:
* The `sort` command sorts the lines of a file alpha-numerically and prints the sorted result to standard output.
* If you enter **`sort -R pets.txt`** you get back the lines sorted in the reverse order, with the **"dog"** lines appearing before the **"cat"** lines.



# Excluding repeated lines from views

**`uniq`**:
* Stands for **Unique**.
* Filtered out repeated lines.
* Note that the `uniq` command **only removes duplicated lines if they are consecutive**.



# Extracting lines matching a pattern

**`grep`**:
* Stands for **global regular expression print**.
* Returns lines of a file matching a specified pattern, such as a regular expression.
* The `-i` option expands the pattern search by making it **case-insensitive**.



# Extracting slices from lines

**`cut`**: Extract specific sections from each line in your file.

You can use the `cut` command to extract the **second to ninth characters from every line** by entering **`cut -C 2-9 people.txt`**.



# Extracting field from lines

**`paste`**: Merge lines from multiple different files.
* Note that the columns of each line of the files are evenly combined with a **tab** as the **default delimiter**.
* You can **specify a delimiter other than Tab** using the `-D` option with the `paste` command.

# Recap

You learned how to:
* View the lines of a file alpha-numerically using sort.
* Remove repeated lines from your view with unique.
* Get the lines of a file that match your desired criteria using grep.
* Extract slices and fields from lines with cut.
* Merge lines from different files using paste.
