# Listing your directory contents

**`ls`**: List files and directories.

**`ls -l`**: List files and directories in more detailed format.

You can see that the terminal has listed all the child files, directories, and additional details, such as:
* permissions, 
* last-modified date, 
* and owner.





# Finding your working directory

**`pwd`**: print current working directory.



# Navigating your directories

**`cd`**: change current directory.



# Relative and absolute navigation

The `cd` command enables you to change directories with either a **relative** or an **absolute** path.



# Finding files

**`find`**: find files in directory tree.

The **`find`** command is a powerful tool that will return the path to every file that matches a user-specified criterion.



**`find . -name a.txt.`**
* The **`.`**(dot) argument means **search in current directory**, so the command will only search within your current working directory and its sub-directories.
* To perform a case-insensitive version of your search, you can use the **`-iname`** option instead.

# Recap

You learned that:
* The `ls` command lists all files and directories contained within a specified directory tree
* The `cd` command is used to navigate between directories.
* Relative paths are relative to your current working directory, while absolute paths stand independently.
* The `find` command is used to find files in your directories.
