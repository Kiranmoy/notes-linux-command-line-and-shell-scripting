# Overview of the Linux Shell

The Linux shell is an OS-level application that interprets commands.

You can use shell commands to perform tasks such as:
* moving and copying files, 
* writing to and reading from files, 
* extracting and filtering data, 
* and searching for data.

There are many types of shells, but the base functionality of most is the same.

Some popular examples include **Bash** and **Zsh**.

# Overview of a Linux Terminal

A terminal is an application, or user interface, where you enter the commands you want to run and receive any output from those commands.
* When you press **Enter**, the terminal passed the commands & arguments to the shell.
* The shell executes the command and returns the result to the terminal.
* The terminal shows the result to the user.

# Communicating with the Linux system



**How are commands run?**

* First, we have a user who wants to run a command.
* They enter the command in a terminal, which is then relayed to the shell.
* The core components of the operating system and kernel translate the command for the hardware to perform.
* When the hardware completes the command, the kernel reads any changes or results and sends them back via the shell to the terminal for the user’s information.
* The terminal is a powerful way to run applications and interact with your machine.

# The Linux terminal



* The area where you enter commands is called the command line.
* The vertical line, or cursor, is the command prompt.
* This indicates where the text that you type will be displayed.
* Here, the **current working** directory is `/home/me/Documents`.
* Not all terminals display the **full location**, or **path**, of your current directory.

# Path in the Linux file system

The path is the human-readable location of a directory or file in the Linux filesystem, for example: `/home/me/Documents`.

There are also special paths:
* **`~`**: User’s Home directory
* **`/`**: Root directory
* **`..`**: Two periods refer to the parent of the current directory
* **`.`**: Current directory

# Changing the current working directory

```bash
cd /                     # change to root directory
cd bin                   # change to bin directory
ls                       # listing directory

cd ..                    # goto parent directory
cd ~                     # goto USER home directory
```
