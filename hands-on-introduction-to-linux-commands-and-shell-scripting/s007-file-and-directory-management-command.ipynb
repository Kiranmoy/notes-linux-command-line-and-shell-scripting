**`mkdir`**: Make directory.



# Removing files and directories

**`rm`**: remove files or directory.



You cannot simply remove **“folder1”** since it may contain other files. However, you can easily get around this by entering `rm` with the `-r` option.
* The `-r` option means you want to remove the directory along with all its child file objects. Your Documents folder is now empty.
* Accordingly, you should always be careful when using `rm` with the `-r` option. It is very easy to accidentally remove folders with important data.

Now, suppose you create an empty directory with the `mkdir` command, and then you decide to remove it. 
* The `rm -rf` command is not recommended.
* Instead, you should use the rmdir command, which is used solely to remove empty directories. This guarantees that you’ll never accidentally delete any important files or directories.

# Creating files

**`touch`**
* If the file does not exist, then create a new file.
* If a file exists, then the `touch` command updates the **last modified date** to the current datetime.
